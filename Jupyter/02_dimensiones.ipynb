#### **Generando un vector aleatorio de 12 numeros enteros (deade 1 al 10) con random**

```python
vector = np.random.randint(1, 11, 12)
vector
```

#### **obtencion de la dimension de un arreglo**

```python
np.shape(vector)
```

#### **Transformar el vector de 12 elementos en una matriz 4x3**
El m√©todo reshape() redimensiona una estructura

```python
matriz = vector.reshape(4,3)
matriz
```

#### **otra variante de matriz (6x2) genrada a partir del vector**

```python
matriz2 = vector.reshape(6,2)
matriz2
```

#### **volviendo a consultar la dimension pero en la matriz**

```python
np.shape(matriz)
```

#### **Conociendo la biblioteca matplotbib**
el paquete matplotlib permite  crear vizualizaciones (graficos) para interpretar resultados y data.

```python
import matplotlib.pyplot as plt
```

### **creando un grafico simple**

```python
plt.plot(vector, label = "vector")
plt.title("Grafico del Vector")
plt.xlabel("indice") # agrega el titulo del eje x
plt.ylabel("valores") # agrega el titulo del eje y

plt.grid(True) # a√±adir una grilla al grafico 
plt.legend("vector") # etiqueta que se obtiene del label del metodo plot(vector, label = "vector")

plt.show() #permite visualizar graficos 
```

#### **Creando dos vectores aleatorios en el mismo grafico**

```python
vector1 = np.random.randint(1,11, 12)
vector2 = np.random.randint(2,20, 12) # se crean dos plot para tener dos vectores en el mismo grafico 

plt.plot(vector1, marker = "d", label = "vector 1")
plt.plot(vector2, marker = "d", label = "vector 2")
# la propiedad marker hace una marca donde hay cada punto

plt.title("Grafico de dos Vectores")
plt.xlabel("indice") 
plt.ylabel("valores") 

plt.grid(True) 
plt.legend("vector") 

plt.show() 
```

#### **Creando un grafico en barras**

```python
vector1 = np.random.randint(1,11, 12)
vector2 = np.random.randint(2,20, 12) # se crean dos plot para tener dos vectores en el mismo grafico 

plt.bar(range(len(vector2)), vector1)
# al cambiar plot a bar es un grafico de barras

plt.title("Grafico de dos Vectores")
plt.xlabel("indice") 
plt.ylabel("valores") 

plt.grid(True) 
plt.legend("vector") 

plt.show() 
```
