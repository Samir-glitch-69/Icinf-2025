##### **Importando diferentes bibliotecas**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

##### **Cargando el dataset y creando dataset con cantidad especificada**

```python
ruta = "data\Titanic-Dataset.csv"
#idealmente el df es para 'dataframe' que es para definir diferentes dataframes
#el id es lo que se basa el index es de orden de menor a mayor siempre 
df = pd.read_csv(ruta, index_col='PassengerId')
#adentro del head se coloca nada para que muestre solamente 5 y si se coloca un numero se muestra la cantidad
#de pasajeros etc
df.head()
```

#### **Informacion del dataset**
se refiere a la informacion basica como filas columnas y otros datos

```python
print(f'Cantidad de Filas y Columnas del Dataset: {df.shape}\n')
df.info()
print()
df.describe()
```

#### **consultando y eliminando duplicados**

```python
duplicados = df.duplicated().sum()
if duplicados > 0:
    print(f"se encontraron {duplicados} filas duplicadas. procediendo a eliminarlas")
    df.drop_duplicates(inplace=True)
else:
    print("no se encontraron filas duplicadas no es necesario eliminarlas")

print("dimensiones tras el paso de duplicados:", df.shape)
```

#### **Identiicando y eliminando nulos**

```python
#cuantos nulos hay en cada columna
nulos_columnas = df.isnull().sum()
#nulos en filas
nulos_filas = df[df.isnull().any(axis=1)]
#suma el total de nulos
total_nulos = nulos_columnas.sum()


nulos_columnas
```
